```python
import numpy as np
import matplotlib.pyplot as plt
from ripser import Rips
import persim
```

# Creating the Dataset

```python
# Number of data points
num_points = 100

# Generate angles evenly spaced around the circle
angles = np.linspace(0, 2 * np.pi, num_points)

# Creates the circle: points on a circle of radius 1
x = np.cos(angles)
y = np.sin(angles)

# Stack coordinates into a 2D array (each row is a point)
data = np.vstack((x, y)).T

# Add Gaussian noise to each point
noise = np.random.normal(0, 1, data.shape)
noisy_data = data + noise

# plot the noisy circle data
plt.figure(figsize=(6, 6))
plt.scatter(noisy_data[:, 0], noisy_data[:, 1], color='blue', label='Noisy Circle')
plt.title("Synthetic Noisy Circle Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

# Computing Persistent Homology

```python
# Instantiate the Vietoris-Rips solver
rips = Rips(maxdim=1)

# Compute the persistence diagrams for the noisy circle data
diagrams = rips.fit_transform(noisy_data)

print("Persistent Diagram:", diagrams)
```

# Visualizing the Persistence Diagram

```python
# Create a new figure for the persistence diagram
plt.figure(figsize=(6, 6))

# Plot the persistence diagram
persim.plot_diagrams(diagrams, show=True)

plt.title("Persistence Diagram for the Noisy Circle")
plt.xlabel("Birth")
plt.ylabel("Death")
plt.show()
```
